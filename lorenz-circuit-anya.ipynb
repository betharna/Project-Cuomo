## Lorenz Circuit 

1. Introduction

The Lorenz system is a simplified mathematical model derived from the equations governing fluid convection. It originates from the study of Rayleigh–Bénard convection, where a fluid layer is heated from below and cooled from above.

When the temperature gradient exceeds a critical threshold, the system transitions from conduction to convection, producing circulating fluid motion. The full system is described by nonlinear partial differential equations, which are difficult to analyze directly. To address this, a reduced-order model was developed that captures the essential dynamics of the system.

2. Governing Equations of Convection

The motion of a viscous fluid with temperature variation is governed by three equations:

Momentum Equation (Navier–Stokes with buoyancy)

```python
Physical Interpretation of Variables

The variables represent amplitudes of dominant modes:

x: intensity of convective motion
y: horizontal temperature difference
z: vertical temperature distortion

These are not direct measurements but coefficients in a truncated expansion of the physical fields.
```

Lorenz’s Model Reduction

To simplify the system, a low-dimensional approximation is introduced:

Assume two-dimensional convection rolls
Represent velocity and temperature using Fourier series
Retain only the dominant modes

This procedure reduces the infinite-dimensional PDE system to a finite-dimensional system of ODEs using Galerkin truncation.

# Computation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lorenz(state, sigma=10, r=28, b=8/3):
    x, y, z = state
    
    dx = sigma * (y - x)
    dy = r * x - y - x * z
    dz = x * y - b * z
    
    return np.array([dx, dy, dz])
```

```python
def euler_step(f, state, dt):
    return state + dt * f(state)
```

```python
dt = 0.01
steps = 10000

state = np.array([1.0, 1.0, 1.0])
trajectory = []

for _ in range(steps):
    state = euler_step(lorenz, state, dt)
    trajectory.append(state)

trajectory = np.array(trajectory)
```

```python
plt.plot(trajectory[:,0], trajectory[:,2])
plt.title("Lorenz Attractor")
plt.show()
```

The figure shows the trajectory of the Lorenz system in phase space The most striking feature is the butterfly shape, which is the classic Lorenz attractor.

The trajectory does not repeat
It stays bounded
It forms two lobes (left and right)

This indicates that the system has evolved onto a:

strange attractor

which is a hallmark of chaotic systems. The plot exhibits a strange attractor characterized by bounded, non-periodic trajectories and sensitive dependence on initial conditions, confirming that the simulated system operates in a chaotic regime.a
